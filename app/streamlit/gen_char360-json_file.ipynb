```python
file_360 = "char360_fib_num.csv"
df = pd.read_csv(file_360)
```

```python
df.head()
```

```python
zi_360 = df.to_dict("records") # df["Zi"].to_list()
```

```python
len(zi_360), zi_360[:10]
```

```python
for z in zi_360:
    zi, fib_num = z["zi"], z["fib_num"]
    fib = str(fib_num).zfill(3)
    file_name = Path("char360-json") / f"{fib}-{zi}-1.json"
    print(str(file_name))
    with open(file_name, "w", encoding="utf-8") as f:
        f.write("{}")
```
